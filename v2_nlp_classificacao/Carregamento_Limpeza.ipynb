```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, PoissonRegressor
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
import matplotlib.dates as mdates



df_items = pd.read_csv('olist_order_items_dataset.csv')
df_products = pd.read_csv('olist_products_dataset.csv')
df_orders = pd.read_csv('olist_orders_dataset.csv')
```

```python
df = pd.merge(df_items, df_products[['product_id', 'product_category_name']], on='product_id', how='left')

df = pd.merge(df, df_orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='left')

df = df.rename(columns={
              'product_category_name': 'item',
              'price': 'sells'})


df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['date_only'] = df['order_purchase_timestamp'].dt.date


df['cleaned itens'] = df['item'].str.lower().str.strip()
df['sells'] = df['sells'].fillna(0)

# Criando o DF de salvamento usando os nomes NOVOS 
df_vendas_limpas = df[['order_id', 'product_id', 'sells', 'item', 'date_only']].copy()

#  Validação 
faturamento_validacao = df_vendas_limpas[df_vendas_limpas['item'] == 'cool_stuff']['sells'].sum()
print(f"Faturamento da categoria validado: R$ {faturamento_validacao:.2f}")

# Preparação para a regressão
df_category = df[df['item'] == 'cool_stuff']
df_sales_day = df_category.groupby('date_only')['sells'].sum().reset_index()

# Salvar os arquivos
df_vendas_limpas.to_csv('vendas_consolidadas.csv', index=False)
```

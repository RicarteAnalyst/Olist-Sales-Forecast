```python
import joblib

#Carregando o 'conhecimento' salvo
model = joblib.load('modelo_sentimento.pkl')
vetorizador = joblib.load('vetorizador.pkl')

def analisar_comentario(texto):
    # Transformando o texto usando o dicion치rio que o modelo 
    texto_vetorizado = vetorizador.transform([texto])
    
    # Faz a predi칞칚o
    predicao = model.predict(texto_vetorizado)[0]
    probabilidade = model.predict_proba(texto_vetorizado).max()
    
    sentimento = "Positivo 游땕" if predicao == 1 else "Negativo 游땨"
    
    print(f"Coment치rio: {texto}")
    print(f"Sentimento: {sentimento}")
    print(f"Confian칞a do modelo: {probabilidade:.2%}")
    print("-" * 30)

# --- TESTE PR츼TICO ---
feedback = input("Digite um coment치rio para testar a IA: ")
analisar_comentario(feedback)
```
